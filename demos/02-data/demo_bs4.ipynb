# Beautifulsoup 4
HTML Parser

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
r = requests.get("https://en.wikipedia.org/wiki/List_of_districts_of_Karnataka")
```

```python
soup = BeautifulSoup(r.text)
```

```python
print(soup.prettify())
```

```python
print(soup.title)
```

```python
tabs = soup.find_all('table')
```

```python
len(tabs)
```

```python
req_tab = tabs[2]
```

```python
print(req_tab)
```

```python
tbod = req_tab.find_all('tr')
dist_codes ={}
codes_pop = {} 
for trow in range(1,len(tbod)-1):
    tds = tbod[trow].find_all('td')
    code = tds[0].text.strip()
    dist_codes[code] = tds[1].text.strip()
    codes_pop[code] = tds[5].text.strip()
```

```python
print(dist_codes)
```

```python
print(codes_pop)
```
